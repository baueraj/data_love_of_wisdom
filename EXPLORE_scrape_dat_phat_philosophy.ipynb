## Getting list of themes

```python
url = 'http://philosophyideas.com/search/themes.asp'

r = requests.get(url)

soup = BeautifulSoup(r.content) 
print(soup.prettify())
```

```python
table = soup.find('div', attrs = {'class': 'txt'})
```

```python
table
```

```python
table.find_all_next('td', attrs = {'class': 'gists'})
```

```python
headers = table.find_all_next('td', attrs = {'class': 'gists'})
```

```python
headers_new = []

for header in headers:
    substr_start = [idx for idx, x in enumerate(str(header)) if x == '>'][1]
    substr_end = str(header).index(' </a')
    domain = str(header)[(substr_start+1):substr_end] #add the number before the domain name
    headers_new.append(domain)
```

```python
headers_new = [f'{i+1}. {j}' for i, j in zip(range(len(headers_new)), headers_new)]
```
